```python
import pandas as pd
import numpy as np
import os
```

## Import Data

```python
#set path of the raw data
get_raw_data_script_file = os.path.join(os.path.pardir,'src','data','get_raw_data.py')
get_raw_data_script_file =os.path.abspath(get_raw_data_script_file)
get_raw_data_script_path = os.path.join(os.path.pardir,'src','data')
get_raw_data_script_path = os.path.join(os.path.pardir,get_raw_data_script_path,'raw')
if not os.path.exists(get_raw_data_script_path):
    os.makedirs(get_raw_data_script_path)

file_path = os.path.join(get_raw_data_script_path, 'GlobalLandTemperaturesbyCity.csv.zip')


```

## Summary Statistics

```python
temps_df.describe()
```

```python
# numerical feature
#centrality measures
print('Mean Temperature : {0:.5}'.format(temps_df.AverageTemperature.mean()),u"\N{DEGREE SIGN}C") # mean\n,
print('Median Temperature : {0:.5}'.format(temps_df.AverageTemperature.median()),u"\N{DEGREE SIGN}C") # median
```

```python
# dispersion measures
print('Min Temperature : {0:.5}'.format(temps_df.AverageTemperature.min()),u"\N{DEGREE SIGN}C") # minimum\n,
print('Max Temperature : {0:.5}'.format(temps_df.AverageTemperature.max()),u"\N{DEGREE SIGN}C") # maximum\n,
print('Temperature range : {0:.5}'.format(temps_df.AverageTemperature.max()  - temps_df.AverageTemperature.min()),u"\N{DEGREE SIGN}C") # range\n,
print('25 percentile : {0:.5}'.format(temps_df.AverageTemperature.quantile(.25)),u"\N{DEGREE SIGN}C") # 25 percentile\n,
print('50 percentile : {0:.5}'.format(temps_df.AverageTemperature.quantile(.5)),u"\N{DEGREE SIGN}C") # 50 percentile\n,
print('75 percentile : {0:.5}'.format(temps_df.AverageTemperature.quantile(.75)),u"\N{DEGREE SIGN}C") # 75 percentile\n,
print('Variance Temperature : {0:.5}'.format(temps_df.AverageTemperature.var()),u"\N{DEGREE SIGN}C") # variance\n,
print('Standard deviation Temperature : {0:.5}'.format(temps_df.AverageTemperature.std()),u"\N{DEGREE SIGN}C") # standard deviation
```

```python
# take N as +ve values and S as -ve values for latitude

temps_df.loc[temps_df.Latitude.str.contains('N'),'Latitude']  = temps_df.Latitude.str.replace('N', '') 
temps_df.loc[temps_df.Latitude.str.contains('S'),'Latitude'] = "-" + temps_df.loc[temps_df.loc[:,'Latitude'].str.contains('S'),'Latitude'].str.replace('S','')
temps_df.loc[:,'Latitude'] = temps_df.loc[:,'Latitude'].astype(float)
```

```python
# plot a scatter graph of average temperature against latitude for specific date

temps_df_lat = temps_df[temps_df.dt == '2012-09-01'].groupby('Country').mean() #find mean of all numerical values in df for each country

temps_df_lat.reset_index().plot.scatter(x = 'Latitude', y = 'AverageTemperature', color = 'c', title = 'Average Temperature for 2012-09-01');
```

## Grouping and Aggregations

```python
# group by and aggregate
temps_df.groupby('Country').agg({'AverageTemperature': 'mean', 'Latitude': max})
```

```python
# more complex aggregations

aggregations = {
    'AverageTemperature' : ['mean', 'median', max, min],
    'Latitude' : [max, min, lambda x: max(x)-min(x)]
    
}
```

```python
temps_df.groupby('Country').agg(aggregations)
```

```python
# find how many cities for each country

cities = temps_df.groupby('Country').City.nunique()
cities = cities.reset_index()
cities
cities[(cities['Country'] == 'United Kingdom') | (cities['Country'] == 'China') | (cities['Country'] == 'Australia') | (cities['Country'] == 'United States')].plot.bar(x='Country', y = 'City', 
rot = 0, figsize = (12,5), title = 'Count of cities recorded for each country', color = ['y', 'r','g','b'], legend = False);
```

## Convert Date Column

```python
temps_df.dt = temps_df.dt.str.replace('-','')
temps_df['Date'] = pd.to_datetime((temps_df.dt).astype(str), format = '%Y%m%d')
temps_df =temps_df.drop(columns ='dt')
```

```python
newcol = (temps_df.Date.dt.year).astype(str).str[0:3] + '0\'s' 
temps_df["Decade"] = newcol
temps_df
    
```

## Pivot Table

```python
#pivot table
temps_df.pivot_table(index = 'Country', columns = 'Decade', values = 'AverageTemperature', aggfunc = 'mean')
```

```python
temps_df.groupby(['Country', 'Decade']).AverageTemperature.mean()
```

```python
#use unstack() to show groupby in tabular form
temps_df.groupby(['Country', 'Decade']).AverageTemperature.mean().unstack()
```
